### 行列の基本性質
---
■ゼロ行列・単位行列  
$ AO = OA = O, AI = IA = A $
  
---
■逆行列
  
行列$ A $ に対して, $ AX = XA = I $ となるような行列$ X $ を行列$ A $の逆行列という．
> 行列には割り算がない.  
> 積の逆演算 $ AB = C $ のとき, $ B = C / A $ などとしたい場合,逆行列を使う.  
  
---
■直交行列
  
転置行列が逆行列と等しいような正方行列をある行列に対する直交行列という．
  
行列$ A $ の転置行列を $ A^t $ とすると，$ A^tA = AA^t = I $ が成立する．

---
■行列式
行列の特徴を表す指標の一つ．スカラーの値をとる．  
逆行列が存在するか確認ができ，逆行列の導出にも用いられる．  
正方行列のみに定義される．   

- 2次  
$   A = \left(
    \begin{array}{cc}
      a_{11} & a_{12} \\
      b_{21} & b_{22} \\
    \end{array}
  \right)
$  
について，  
$ |A| = a_{11}b_{22} - a_{12}b_{21} $  
- 3次  
$   A = \left(
    \begin{array}{ccc}
      a_{11} & a_{12} & a_{13} \\
      b_{21} & b_{22} & b_{23} \\
      c_{31} & c_{32} & c_{33} \\      
    \end{array}
  \right)
$  
について，  
$ |A| = a_{11}b_{22}c_{33} + a_{12}b_{23}c_{31} + a_{13}b_{21}c_{32} - a_{13}b_{22}c_{31} - a_{12}b_{21}c_{33} - a_{11}b_{23}c_{32} $  

---
■対角化
行列$ A $ を対角化するとは，直交行列 $ P $ を持ってきて，対角行列$ D $ について
$$ A = PDP^t $$
とすることを言う．
  
対角行列$ D $ は，直交行列の性質を利用して，
$$ D = P^tAP $$
と置くことができる．

---
■固有値分解  
n×nの正方行列$ A $ に対して，n次のベクトル$ v $ と，スカラー$ \lambda $を利用し，  
$ Av = \lambda v $  
となる$ v, \lambda $ を見つけたい．上記を満たす$ v, \lambda $ をそれぞれ，固有ベクトル，固有値と呼ぶ．

1． $ Av = \lambda v $ を想定して，行列式を計算する  
2． 行列式はゼロになる $ \lambda $ を見つける  
3． $ Av = \lambda v $ と$ \lambda $ を使って，$ v $ を求める

---
■特有値分解
固有値分解を一般化したもの．
  
行列$ A $ に対して，直行行列$U, V $ ，対角行列$ D $ を用いて，$ A = UDV^t $ と分解する．
